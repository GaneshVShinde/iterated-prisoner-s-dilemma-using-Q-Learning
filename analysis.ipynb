###### Q learning which comes with behaviour to win

```python
from prisonersDelima import *
import seaborn as sns
```

```python
def take_action(history):
        #print("in Q")
        if len(history) > 0: 
            pnts=prisonersDelima.evaluate_points(history)
            reward = pnts[1] -pnts[0]
            state=history[-1]
            ai.learn(state,reward)
            choice=ai.do(state)
            return(choice)
        else:
            return('c')
```

###### Fight against Priyanka

priD=prisonersDelima(0)
ai=qlearn.QLearn(epsilon=qEpsilon,lambd=qLambda,alpha=qAlpha)

```python
priD.set_plr_func(take_action)
roundStates=priD.n_round()
finalVals=prisonersDelima.evaluate_points(roundStates)
priD.plot_scoreMap()
```
